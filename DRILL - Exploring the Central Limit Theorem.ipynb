```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reproduce the pop1 and pop2 populations and samples, using numpy's binomial function. 
#Specifically, create two binomially distributed populations with n equal to 10 and size equal to 10000
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 
```

```python
# Letâ€™s make histograms for the two groups. 
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()
```

```python
#The p-value of pop1 should be 0.2 and the p-value of pop2 should be 0.5. 
#Using a sample size of 100, calculate the means and standard deviations of your samples.
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())
```

```python
#p-values for sample 1 and sample 2 given that mean=n*p
print(sample1.mean()/10)
print(sample2.mean()/10)
```

```python
#Increase the size of your samples from 100 to 1000, then calculate the means and standard deviations 
#for your new samples and create histograms for each. 
#Repeat this again, decreasing the size of your samples to 20. What values change, and what remain the same?
sample1 = np.random.choice(pop1, 1000, replace=True)
sample2 = np.random.choice(pop2, 1000, replace=True)
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())
```

```python
#Repeat this again, decreasing the size of your samples to 20. What values change, and what remain the same?
sample1 = np.random.choice(pop1, 20, replace=True)
sample2 = np.random.choice(pop2, 20, replace=True)
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())
```

```python
#Change the probability value (p in the NumPy documentation) for pop1 to 0.3, 
#then take new samples and compute the t-statistic and p-value. 
pop1 = np.random.binomial(10, 0.3, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 
# Letâ€™s make histograms for the two groups. 
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()
```

```python
#Using a sample size of 100, calculate the means and standard deviations of your samples.
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())
```

```python
#compute the t-statistic and p-value
from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

```python
#Then change the probability value p for group 1 to 0.4, and do it again. What changes, and why?
pop1 = np.random.binomial(10, 0.4, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 
# Letâ€™s make histograms for the two groups. 
plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()
#Using a sample size of 100, calculate the means and standard deviations of your samples.
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)
plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 
plt.legend(loc='upper right') 
plt.show()
print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())
#compute the t-statistic and p-value
from scipy.stats import ttest_ind
print(ttest_ind(sample2, sample1, equal_var=False))
```

```python
#The t-value scales the difference between the two groups by the amount of variance in the two samples. 
#High variability in samples can lead to groups with means that look very different, 
#but when we look at the histogram we see that most of the values in the sample groups overlap.

#The p-value associated with a t-test indicates the likelihood of getting a difference this large 
#or larger in the samples if the populations were not different.
#p-value is the probability that the pattern of data in the sample could be produced by random data.
#The smaller the p-value, the more likely the difference we see in the samples meaningfully reflects the populations.

#pop1 = np.random.binomial(10, 0.3, 10000)
#pop2 = np.random.binomial(10,0.5, 10000) 
#Ttest_indResult(statistic=12.257620916368202, pvalue=4.5280233324164785e-26)

#pop1 = np.random.binomial(10, 0.4, 10000)
#pop2 = np.random.binomial(10,0.5, 10000) 
#Ttest_indResult(statistic=5.084240489977581, pvalue=8.529672714166992e-07)

#Therefore increasing our probability value, p, decreases the T-value and increases the P-value.  So the amount
#of variance is decreasing but since the p-value is increasing, the less likely the deifferences we see in the sample 
#meaningfully reflects the population.
```

```python
#Change the distribution of your populations from binomial to a distribution of your choice. 
#Do the sample mean values still accurately represent the population values?
popy = np.random.geometric(p=0.2, size=10000)
popz = np.random.geometric(p=0.5, size=10000)
plt.hist(popy, alpha=0.5, label='Population Y') 
plt.hist(popz, alpha=0.5, label='Population Z') 
plt.legend(loc='upper right') 
plt.show()
```

```python
sampley = np.random.choice(popy, 100, replace=True)
samplez = np.random.choice(popz, 100, replace=True)
plt.hist(sampley, alpha=0.5, label='sample y') 
plt.hist(samplez, alpha=0.5, label='sample z') 
plt.legend(loc='upper right') 
plt.show()
print(sampley.mean())
print(samplez.mean())
print(sampley.std())
print(samplez.std())
#compute the t-statistic and p-value
from scipy.stats import ttest_ind
print(ttest_ind(samplez, sampley, equal_var=False))
```

```python
#The larger the absolute value of the t-value, the smaller the p-value, 
#and the greater the evidence against the null hypothesis.
```
